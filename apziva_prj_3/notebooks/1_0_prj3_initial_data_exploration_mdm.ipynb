# Reading the Data

```python
import numpy as np
import pandas as pd

data = pd.read_csv("potential-talents - Aspiring human resources - seeking human resources.csv")
data.head()
```

# Understanding the Data

## Shape and Data Types

```python
# checking shape
print(data.shape)
```

```python
# chacking data types
for col in data:
  print(type(data[col][1]))
```

```python
# chacking categorical values
cat = data.select_dtypes(include='O').keys()
cat
```

## Count of each element in dataframe

```python
# count of each element in dataframe
print(data['job_title'].value_counts(), "\n\n")
print(data['location'].value_counts(), "\n\n")
print(data['connection'].value_counts(), "\n\n")
```

## Checking for Missing Values

```python
# checking for missing values
data.isnull().sum()
```

## Cleaning the Data



```python
# changing all uppercase letters to lowercase
data['job_title'] = data['job_title'].str.lower()
data['location'] = data['location'].str.lower()
data['connection'] = data['connection'].str.lower()
data.head()
```

```python
# getting rid of punctuation
data = data.replace('[,\.\|\()\!]','', regex=True)
data.head()
```
